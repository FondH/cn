```python
import matplotlib.pyplot as plt
import pandas as pd
#Part--1
data = {
    "Loss Rate (%)": ["0", "5", "10", "15", "20"],
    "Default - Total Bytes (bs)": [5904369, 6232369, 6281585, 6789969, 7347793],
    "Default - Duration (s)": [17.807, 21.276, 22.357, 27.356, 35.936],
    "Default - Speed Rate (Bps)": [347315, 296779, 285526, 251480, 209936],
    "GBN - Total Bytes (bs)": [5920464, 8806880, 12480464, 13792464, 17416880],
    "GBN - Duration (s)": [16.91, 25.172, 35.8, 39.621, 50.256],
    "GBN - Speed Rate (Bps)": [370029, 352275, 356584, 353652, 348337],
    "SR - Total Bytes (bs)": [5920464, 6166464, 6724064, 7035664, 7232464],
    "SR - Duration (s)": [16.943, 17.098, 17.573, 17.804, 17.569],
    "SR - Speed Rate (Bps)": [370029, 362733, 395533, 413862, 425439]
}

df = pd.DataFrame(data)
df.set_index("Loss Rate (%)", inplace=True)
df
```

```python
loss_rates = df.index.astype(int)  # Converting loss rates to integers for plotting
default_speeds = df["Default - Speed Rate (Bps)"]
gbn_speeds = df["GBN - Speed Rate (Bps)"]
sr_speeds = df["SR - Speed Rate (Bps)"]

plt.figure(figsize=(10, 6))
plt.plot(loss_rates, default_speeds, label='Default', marker='o')
plt.plot(loss_rates, gbn_speeds, label='GBN', marker='s')
plt.plot(loss_rates, sr_speeds, label='SR', marker='^')

plt.xlabel('Loss Rate (%)')
plt.ylabel('Speed Rate (Bps)')
plt.title('Comparison of Speed Rates at Different Loss Rates')
plt.legend()

# Show the plot
plt.show()
```

```python

default_total_bytes = df["Default - Total Bytes (bs)"]
gbn_total_bytes = df["GBN - Total Bytes (bs)"]
sr_total_bytes = df["SR - Total Bytes (bs)"]

plt.figure(figsize=(10, 6))
plt.plot(loss_rates, default_total_bytes, label='Default', marker='o')
plt.plot(loss_rates, gbn_total_bytes, label='GBN', marker='s')
plt.plot(loss_rates, sr_total_bytes, label='SR', marker='^')


plt.xlabel('Loss Rate (%)')
plt.ylabel('Total Bytes Transmitted (bs)')
plt.title('Comparison of Total Bytes Transmitted at Different Loss Rates')
plt.legend()

plt.show()
```

```python
loss_rates = [0, 5, 10, 15, 20]

# GBN data
gbn_data = {
    "Total Bytes (bs)": [5936864, 6904480, 11840880, 12496880, 18433712],
    "Duration (s)": [17.309, 20.152, 34.888, 37.089, 55.435],
    "Speed Rate (Bps)": [349227, 345224, 348261, 337753, 335158]
}

# SR data
sr_data = {
    "Total Bytes (bs)": [5936864, 6396064, 7052048, 7216048, 7954080],
    "Duration (s)": [17.29, 22.167, 27.132, 28.318, 35.745],
    "Speed Rate (Bps)": [349227, 290730, 261186, 257716, 227259]
}


df_gbn = pd.DataFrame(gbn_data, index=loss_rates)
df_sr = pd.DataFrame(sr_data, index=loss_rates)
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# Total Bytes 
axes[0].plot(loss_rates, df_gbn["Total Bytes (bs)"], label='GBN - Total Bytes', marker='o', linestyle='-', color='blue', linewidth=2)
axes[0].plot(loss_rates, df_sr["Total Bytes (bs)"], label='SR - Total Bytes', marker='s', linestyle='--', color='red', linewidth=2)
axes[0].set_title('Total Bytes Transmitted at Different Loss Rates')
axes[0].set_xlabel('Loss Rate (%)')
axes[0].set_ylabel('Total Bytes (bs)')
axes[0].legend()

# Duration 
axes[1].plot(loss_rates, df_gbn["Duration (s)"], label='GBN - Duration', marker='o', linestyle='-', color='blue', linewidth=2)
axes[1].plot(loss_rates, df_sr["Duration (s)"], label='SR - Duration', marker='s', linestyle='--', color='red', linewidth=2)
axes[1].set_title('Duration at Different Loss Rates')
axes[1].set_xlabel('Loss Rate (%)')
axes[1].set_ylabel('Duration (s)')
axes[1].legend()

# Speed
axes[2].plot(loss_rates, df_gbn["Speed Rate (Bps)"], label='GBN - Speed Rate', marker='o', linestyle='-', color='blue', linewidth=2)
axes[2].plot(loss_rates, df_sr["Speed Rate (Bps)"], label='SR - Speed Rate', marker='s', linestyle='--', color='red', linewidth=2)
axes[2].set_title('Speed Rate at Different Loss Rates')
axes[2].set_xlabel('Loss Rate (%)')
axes[2].set_ylabel('Speed Rate (Bps)')
axes[2].legend()

plt.tight_layout()
plt.show()
```
